```python
from pyspark.sql import SparkSession
```

```python
spark=SparkSession.builder \
.master('spark://spark-master:7077') \
.appName('Marketing Analytics') \
.config('spark.driver.memory', '512M')  \
.config('spark.executor.memory', '512M')  \
.config('spark.executor.instances', '1')  \
.config('spark.executor.cores', '1')  \
.config("spark.cores.max", "1") \
.getOrCreate()
```

```python
file_location = "/spark-data/marketing_data.csv" 
file_type = "csv"

infer_schema = "true" 
first_row_is_header = "true" 
delimiter = ","
```

```python
marketingDF = spark.read.format(file_type).option("inferSchema", infer_schema)\
.option("header", first_row_is_header)\
.option("sep", delimiter)\
.load(file_location)

marketingDF.show()
```

```python
from pyspark.sql.functions import col, regexp_replace 
from pyspark.sql.types import DoubleType
```

```python
marketingDF = marketingDF.withColumnRenamed(" Income ", "Income")
```

```python
marketingDF = marketingDF.withColumn("Income", regexp_replace(col("Income"), "[$,]", ""))
```

```python
marketingDF = marketingDF.withColumn("Income", col("Income").cast(DoubleType()))

marketingDF.show(4)
```

```python
import matplotlib.pyplot as plt
```

```python
plt.hist(marketingDF.select('Income').rdd.flatMap(lambda x: x).collect(), bins=30, color='skyblue', edgecolor='black') 
plt.xlabel('Income') 
plt.ylabel('Count') 
plt.title('Income Distribution') 
plt.show()
```
